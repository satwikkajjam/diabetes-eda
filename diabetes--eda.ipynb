```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/Users/sathwik/Downloads/diabetes.csv')
```

```python
print("ðŸ“Š Summary Statistics:")
display(df.describe())
```

```python
plt.figure(figsize=(12, 8))
for i, col in enumerate(df.columns[:-1]):  # Exclude 'Outcome'
    plt.subplot(3, 3, i+1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.savefig('/Users/sathwik/Downloads/distributions.png')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop('Outcome', axis=1))
plt.title("Boxplot of Features (Outlier Check)")
plt.xticks(rotation=45)
plt.savefig('/Users/sathwik/Downloads/boxplots.png')
plt.show()
```

```python
corr = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix")
plt.savefig('/Users/sathwik/Downloads/correlation_matrix.png')
plt.show()
```

```python
sns.pairplot(df.sample(100), hue='Outcome', diag_kind='kde')
plt.savefig('/Users/sathwik/Downloads/pairplot.png')
plt.show()
```
